```python
import pandas as pd
import requests
import numpy as np
from bs4 import BeautifulSoup
import csv
job_list=[]
link_list=[]
top2=[]
top_other=[]
it_skills_list=[]
skill_by_category=[]
by_category_key=[]
by_category_value=[]

for page_number in range (1,2):

    main_url="http://www.itjobswatch.co.uk"
    my_url = ("https://www.itjobswatch.co.uk/default.aspx?page={}".format(page_number) + "&sortby=0&orderby=0&q=&id=0&lid=2618")
        
    page = requests.get(my_url).text
    soup=BeautifulSoup (page, "html.parser")
    
    #JOB LIST
    for job in soup.findAll("table", attrs={"class":"results"}):
        for kk in job.findAll("td", "c2"):
            job_list.append(kk.text)
    
    #LINK LIST 
    for aa in soup.findAll("td", "c2"):
        for bb in aa.findAll("a"):
            link_list.append(main_url + bb.get("href"))        

    #TOP 30 LIST (IN EVERY SUB-PAGE)
    #for linknumber in range(0, len(link_list)):

    for linknumber in range(0, 1):
      
        top_30=[]
        it_skills_list=[]
        skill_set_title=[]
        skill_set_text=[]
        skill_set_value=[]
        skill_by_category=[]
        by_category_key=[]
        by_category_value=[]
        tt=[]
        tv=[]

        page1 = requests.get(link_list[linknumber]).text
        #print(link_list[linknumber])

        soup1=BeautifulSoup (page1, "html.parser")
        
        ########## TOP 30 
        
        allinfo= soup1.findAll("table", attrs={"class":"tab"})
        for aa in allinfo[1].findAll("td", attrs={"class":"tabRelatedGroupsContainer"}):
            for bb in aa.findAll("td", "cd"):
                for cc in bb.findAll("a"):
                    tt.append(cc.text) # Top 30 texts
            
            for dd in aa.findAll("td", "cp"):
                tv.append(dd.text) # Top 30 values
                
        for i in range(0, len(tt)):
            top_30.append([tt[i], tv[i]])
        #print(top_30)
        
        # SKILL SETS
        for aa in allinfo[2].findAll("td", attrs={"class":"tabRelatedGroupsContainer"}):
            for qq in range(3,22):
                for aa[qq] in aa.findAll ("th", attrs={"colspan":"3"}):#Skill set titles
                    skill_set_title.append(aa[qq].text)
                    for bb in aa.findAll("td", "cd"):
                        for cc in bb.findAll("a"): #Skill set texts
                            skill_set_text.append(cc.text) # Skill set by category


                    for dd in aa.findAll("td", "cp"):
                        skill_set_value.append(dd.text) # Top 30 values

        for i in range(0, len(skill_set_title)):
            for j in range (0, len(skill_set_text)):
                skill_by_category.append([skill_set_title[i], skill_set_text[j], skill_set_value[j]])

        print(skill_by_category)
    
```
